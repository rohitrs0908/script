```python
#import boston dataset from sklearn library
from sklearn.datasets import load_boston
#import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#load dataset
boston_real_state_dataset = load_boston()
```

```python
#view Boston dataset
print(boston_real_state_dataset.DESCR)
```

```python
#define X axis from data
x_axis = boston_real_state_dataset.data
#x_axis
```

```python
#define Y axis from target
y_axis = boston_real_state_dataset.target
#y_axis
```

```python
#plot histogram
style.use('ggplot')
plt.figure(figsize=(5,5))
plt.hist(y_axis,bins=50) # plot y_axis , no. of bins=50
plt.xlabel('Price in 1000 USD')
plt.ylabel('Number of Houses')
plt.show()
```

```python
#view shape
boston_real_state_dataset.data.shape
```

```python
#view column names
boston_real_state_dataset.feature_names
```

```python
#plot scatterplot
style.use('ggplot')
plt.figure(figsize=(5,5))
plt.scatter(x_axis[:,5],y_axis)  # x_axis 6th column vs y_axis 14th column in dataset
plt.xlabel('Price in 1000 USD')
plt.ylabel('Number of Houses')
plt.show()
```
